# 조건문

## if

```python
switch = True
if switch: # if (...): 조건식 -> 최종적으로 bool형태로 표현될 수 있는 값
  print("아 밝다") # if 본문.
```

```python
temperature = 40
if temperature > 38:
  print("아 덥다")
```

```python
temperature = 30
if temperature > 38:
  print("아 덥다")
  print("물 마시고 싶어") 
```

## else

```python
temperature = 30
if temperature > 38:
  print("아 덥다")
else: # temperature <= 38:
  print("조아")
```

## pass

```python
temperature = 30
if temperature > 38:
  print("아 덥다")
else: # temperature <= 38:
  pass # block 유지 때문에
```

## while

```python
for i in range(10):
  print(i, end=" ")
```

```python
i = 0 # 초기식 
while i < 10: # while 조건식
  print(i, end= " ") # ...
  i += 1 # 변화식
```

```python
money = 100000
hit = 0
while money > 0:
  money -= 10
  hit += 1
  print(money, "원 남았습니다", hit, "대 맞았습니다")
```

## break, continue

```python
for i in range(1000):
  print(i, i ** 2)
  if i ** 2 > 1000:
    break
```

```python
for i in range(1, 1000):
  if i % 2 == 0:
    continue
  print(i, end =' ')
```

```python
for i in range(1, 1000):
  if i % 3 != 0:
    continue
  print(i, end =' ')
```

```python
i = 0
while True:
  i == 1
  print(i)
  if i > 10:
    break
```
